```python
#문제 1번
import csv
f= open('./high_school_2019.csv', 'r', encoding='utf8')
lines = csv.reader(f)

header = next(lines) 
#print(header)
result = []
for line in lines:
    result.append(line[:]) 
print(result[:3])
f.close()
```

```python
#2번
import numpy as np

data=np.array(result)

print('-'*90)
for j in range(len(header)-2):
    print(header[j+2],end='')
print()
print('-'*90)
for ix in range(len(data)):
    print(data[ix][2:])
```

```python
print(data)
```

```python
#3번
total_data=np.delete(data,(0,1,2,4,6),axis=1) #불필요 열 제거
total_data=total_data.astype('int64') # 학생수 합산을 위한 int형변환
del_school=np.delete(data,(0,1),axis=1) #결과값 출력을 위한 data

#del_school과 데이터 합치기 위한 차원변환
add_data=total_data.sum(axis=1).reshape(-1,1)
result=np.concatenate([del_school,add_data],1) #데이터 결합
result=result.astype('int64')
print('-'*70)
for j in range(len(header)-2):
    print(header[j+2],end='')
print(" 총학생수")
print('-'*70)
print(result)
```

```python
print(data)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

student_total=np.delete(data,(0,1,2,4,6),axis=1)
student_total=student_total.astype('int')
total_list=student_total.sum(axis=0)
xlabel=['1_total','2_total','3_total']
print("1학년 총 학생 수 : {0}".format(total_list[0]))
print("2학년 총 학생 수 : {0}".format(total_list[1]))
print("3학년 총 학생 수 : {0}".format(total_list[2]))
plt.bar(xlabel,total_list)
plt.title('total_student')
plt.show()
```

```python
#14장 연습문제
#1번
import csv
f= open('./population_age_2020.csv', 'r', encoding='utf8')
lines = csv.reader(f)

header = next(lines) 
#print(header)
result = []
for line in lines:
    result.append(line[:]) 
print(result[:4])
f.close()
```

```python
print(header)
```

```python
#2번
result_array=np.array(result)
transpose_result=result_array.T #전치행렬
print(transpose_result[:4])
```

```python
#3번
dict_shape=dict(zip(header,transpose_result)) #사전형 변환
print(dict_shape)
```

```python
#4번
import pandas as pd
df=pd.DataFrame(dict_shape)
df
```
