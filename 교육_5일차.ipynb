```python
class Invalid_A(Exception):
    def __init__(self,arg):
        super().__init__('시작지점({0})이 잘못되었습니다. '.format(arg))

class Invalid_B(Exception):
    def __init__(self,arg):
        super().__init__('인덱스범위를 벗어났습니다. '.format(arg))

def midstr(s,a,b):
    
    fin=a+b #시작지점부터 끝나는지점 slicing을 위한 변수

    if a > len(s): #시작지점이 문자열 개수를 초과할 때 ex) 'abcde' 시작지점:6
        raise Invalid_A(a)
    elif fin-1 > len(s): #종료지점이 문자열 개수를 초과할 때
        raise Invalid_B()
    else:
        result=s[a-1:fin-1]
        return "결과 문자열은 {0} 입니다".format(result)


s=input("대상 문자열 입력: ")
a=int(input("시작 지점 입력: "))
b=int(input("취할 갯수 입력: "))    
try:
    result=midstr(s,a,b)
except Invalid_A as err:
    print('예외가 발생했습니다: {0}'.format(err))s
except Invalid_B as err:
    print('예외가 발생했습니다: {0}'.format(err))
else:
    print("결과문자열은 {0}입니다".format(result))
```
